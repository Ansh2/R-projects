```r
first_digit <- function(number) {
    number <- abs(number)
    if (number != 0) {
      while (number >= 10) {
        number <- number/10
      }
      while (number < 1) {
        number <- number*10
      }  
      number <- floor(number)
    }        
    return(number)

}
```

```r
first_digit(-6384.52)
```

```r
first_digit(7223.0026)
```

```r
first_digit(0.0037291)
```
